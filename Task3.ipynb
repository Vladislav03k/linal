__4.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
0 & 3 & 0\\ 
3 & 0 & 0\\
0 & 0 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(3, -3, -4)$ собственным вектором этого линейного оператора.

```python
import numpy as np
```

```python
a = np.array([[-1, -6], [2, 6]])
v, w = np.linalg.eig(a)

print(f'Матрица A:\n{a}')
print(f'Собственные значения:\n{w}')
print(f'Собственные векторы:\n{v}')
```

Задача 2:

```python
a = np.array([[-1, 0], [0, -1]])
v, w = np.linalg.eig(a)

print(f'Матрица A:\n{a}')
print(f'Собственные значения:\n{w}')
print(f'Собственные векторы:\n{v}')
```

Получается оператор у нас еденичный, значит и матрицы у нас будет всегда в еденичном базисе и отталкиваясь от этого любой вектор будет собстсвенным для оператора

Задача 3:

```python
a = np.array([[1, 1], [-1, 3]])
v, w = np.linalg.eig(a)

print(f'Матрица A:\n{a}')
print(f'Собственные значения:\n{w}')
print(f'Собственные векторы:\n{v}')
```

При векторе (1, 1) значения сигмы получаются 0 и -4, значит такая система не имеет значения

Задача 4:

```python
a = np.array([[0, 3, 0], [3, 0, 0], [0, 0, 3]])
v, w = np.linalg.eig(a)

print(f'Матрица A:\n{a}')
print(f'Собственные значения:\n{w}')
print(f'Собственные векторы:\n{v}')
```

По подсчетам вектор (3, -3, -4) также является собсвтенным при сигма = -3
