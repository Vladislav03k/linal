```python
import numpy as np
```

```python
from numpy.linalg import det
```

```python
A = np.array(((1, -2), (3, -4)))
A1 = np.array(((1, 7), (3, -4)))
A2 = np.array(((1, -2), (1, 7)))
```

```python
print(det(A))
```

Определитель матрицы не равен 0, значит система совместна

```python
x1 = det(A1)/det(A)
x2 = det(A2)/det(A)
print(x1, x2)
```
