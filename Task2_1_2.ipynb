__5*.__ Написать на Python функцию для перемножения двух произвольных матриц, не используя NumPy.

```python
def matrix_multiply(A, B):
    m_1 = len(A)
    n_1 = len(A[0])
    m_2 = len(B)
    n_2 = len(B[0])

    s = 0  # сумма
    t = []  # временная матрица
    C = []  # конечная матрица

    if n_1 != m_2:
        print('Impossible')
    else:
        for z in range(0, m_1):
            for j in range(0, n_2):
                for i in range(0, n_1):
                    s += A[z][i] * B[i][j]
                t.append(s)
                s = 0
            C.append(t)
            t = []
    return C

A = [[4, 1], [5, -2], [2, 3]]
B = [[4,  5, 2], [1, -2,  3]]
print("AxB\n",matrix_multiply(A, B))
```
